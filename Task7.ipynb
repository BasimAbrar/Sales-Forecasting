# Merging Train + Features + Stores

```python
trainingData = pd.merge(df3, df1, on=["Store", "Date"], how="left")
trainingData = pd.merge(trainingData, df2, on="Store", how="left")
if "IsHoliday_y" in trainingData.columns:
    trainingData.drop(columns=["IsHoliday_y"], inplace=True)
print("Train shape:", trainingData.shape)
```

```python
trainingData.head()
```

# Merging Test + Features + Stores

```python
testingData = pd.merge(df4, df1, on=["Store", "Date"], how="left")
testingData = pd.merge(testingData, df2, on="Store", how="left")
if "IsHoliday_y" in testingData.columns:
    testingData.drop(columns=["IsHoliday_y"], inplace=True)
print("Test shape:", testingData.shape)
```

```python
testingData.head()
```

```python
print("Missing values in trainingData:\n", trainingData.isnull().sum())
print("\nMissing values in testingData:\n", testingData.isnull().sum())
```

# Handling Missing Values

```python
for col in ["Temperature", "Fuel_Price", "CPI", "Unemployment"]:
    trainingData[col] = trainingData[col].ffill().bfill()
    testingData[col] = testingData[col].ffill().bfill()
```

```python
for col in ["MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"]:
    trainingData[col] = trainingData[col].fillna(0)
    testingData[col] = testingData[col].fillna(0)
```

```python
# Rename IsHoliday_x -> IsHoliday
trainingData.rename(columns={"IsHoliday_x": "IsHoliday"}, inplace=True)
testingData.rename(columns={"IsHoliday_x": "IsHoliday"}, inplace=True)
```

```python
print("Missing values in trainingData:\n", trainingData.isnull().sum())
print("\nMissing values in testingData:\n", testingData.isnull().sum())
```

# Feature Engineering
______________________________________________________________________________________________________________________

**Extracting Date-Based Features:**

```python
trainingData["Year"] = trainingData["Date"].dt.year
trainingData["Month"] = trainingData["Date"].dt.month
trainingData["WeekOfYear"] = trainingData["Date"].dt.isocalendar().week.astype(int)
trainingData["DayOfWeek"] = trainingData["Date"].dt.dayofweek
trainingData["IsWeekend"] = trainingData["DayOfWeek"].apply(lambda x: 1 if x >= 5 else 0)

testingData["Year"] = testingData["Date"].dt.year
testingData["Month"] = testingData["Date"].dt.month
testingData["WeekOfYear"] = testingData["Date"].dt.isocalendar().week.astype(int)
testingData["DayOfWeek"] = testingData["Date"].dt.dayofweek
testingData["IsWeekend"] = testingData["DayOfWeek"].apply(lambda x: 1 if x >= 5 else 0)

print(trainingData[["Date","Year","Month","WeekOfYear","DayOfWeek","IsWeekend"]].head(10))
```

```python

print(trainingData[["Date","Year","Month","WeekOfYear","DayOfWeek","IsWeekend"]].head(10))
```

**Creating Lag Features:**

```python
trainingData = trainingData.sort_values(["Store", "Dept", "Date"])
trainingData["Lag_1"] = trainingData.groupby(["Store", "Dept"])["Weekly_Sales"].shift(1)
trainingData["Lag_2"] = trainingData.groupby(["Store", "Dept"])["Weekly_Sales"].shift(2)
trainingData["Lag_3"] = trainingData.groupby(["Store", "Dept"])["Weekly_Sales"].shift(3)
```

**Creating Rolling Average Features (Bonus):**

```python
trainingData["Rolling_Mean_3"] = (
    trainingData.groupby(["Store", "Dept"])["Weekly_Sales"]
    .shift(1)
    .rolling(window=3, min_periods=1)
    .mean()
)

trainingData["Rolling_Mean_12"] = (
    trainingData.groupby(["Store", "Dept"])["Weekly_Sales"]
    .shift(1)
    .rolling(window=12, min_periods=1)
    .mean()
)

trainingData = trainingData.dropna(subset=["Lag_1", "Lag_2", "Lag_3", "Rolling_Mean_3", "Rolling_Mean_12"])
```

```python

print(trainingData[["Date","Store","Dept","Weekly_Sales","Lag_1","Lag_2","Lag_3","Rolling_Mean_3","Rolling_Mean_12"]].head(20))

```

Since test.csv has no Weekly_Sales, we can’t create lags or rolling averages there directly.

# Seasonal Decomposition (Bonus)
______________________________________________________________________________________________________________________

```python
from statsmodels.tsa.seasonal import seasonal_decompose

weekly_sales = trainingData.groupby("Date")["Weekly_Sales"].sum()
weekly_sales = weekly_sales.sort_index()
decomposition = seasonal_decompose(weekly_sales, model="additive", period=52)
plt.rcParams.update({"figure.figsize": (12,8)})
decomposition.plot()
plt.show()
```

# Model Training
______________________________________________________________________________________________________________________

**Linear regression:**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
features = [
    "Store", "Dept", "Size", "Temperature", "Fuel_Price",
    "CPI", "Unemployment", "Year", "Month",
    "Lag_1", "Lag_2", "Lag_3", "Rolling_Mean_3"
]

X = trainingData[features]
y = trainingData["Weekly_Sales"]
```

```python
# Manual Time-Aware Train-Test Split (80% train, 20% validation)
split_index = int(len(X) * 0.8)

X_train, X_val = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_val = y.iloc[:split_index], y.iloc[split_index:]

print(f"Train shape: {X_train.shape}, Validation shape: {X_val.shape}")
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_val)
```

```python
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)
```

```python
print("Linear Regression Performance: ")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
plt.figure(figsize=(12,5))
plt.plot(y_val.values[:100], label="Actual Sales", marker="o")
plt.plot(y_pred[:100], label="Predicted Sales", marker="x")
plt.title("Linear Regression - Actual vs Predicted Sales")
plt.xlabel("Sample Index")
plt.ylabel("Weekly Sales")
plt.legend()
plt.show()
```

**XGBoost Regressor**


```python
from xgboost import XGBRegressor
```

```python
features = [
    "Store", "Dept", "Size", "Temperature", "Fuel_Price",
    "CPI", "Unemployment", "Year", "Month", "WeekOfYear", "DayOfWeek", "IsWeekend",
    "Lag_1", "Lag_2", "Lag_3", "Rolling_Mean_3", "Rolling_Mean_12"
]

X = trainingData[features]
y = trainingData["Weekly_Sales"]
```

```python
# Manual Time-Aware Split (80-20)
split_index = int(len(X) * 0.8)
X_train, X_val = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_val = y.iloc[:split_index], y.iloc[split_index:]
```

```python
xgb_reg = XGBRegressor(
    n_estimators=500,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
xgb_reg.fit(X_train, y_train)
y_pred = xgb_reg.predict(X_val)
```

```python
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)

print("XGBoost Regressor Performance:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
plt.figure(figsize=(12,5))
plt.plot(y_val.values[:100], label="Actual Sales", marker="o")
plt.plot(y_pred[:100], label="Predicted Sales", marker="x")
plt.title("XGBoost - Actual vs Predicted Sales")
plt.xlabel("Sample")
plt.ylabel("Weekly Sales")
plt.legend()
plt.show()
```
